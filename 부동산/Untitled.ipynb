```python
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')

import matplotlib 

matplotlib.font_manager._rebuild()

plt.rc("font", family="Malgun Gothic")
```

```python
fig = plt.figure(figsize=(30,7))
sns.scatterplot(x = unsold['Unnamed: 0'], y = unsold['서울[호]'], color = 'pink', alpha = 0.5 )
plt.plot(unsold['울산[호]'],'--o', color = 'blue', alpha = 0.5)
plt.plot(unsold['부산[호]'],'--o', color = 'green', alpha = 0.5)
plt.plot(unsold['광주[호]'],'--o', color = 'red', alpha = 0.5)
plt.plot(unsold['대구[호]'],'--o', color = 'orange', alpha = 0.5)
plt.plot(unsold['대전[호]'],'--o', color = 'yellow', alpha = 0.5)
plt.xticks(rotation = 90)
plt.show()
```

```python
plt.figure(figsize=(20,3))
sns.barplot(data = unsold, x = 'Unnamed: 0', y = '서울[호]')
plt.tight_layout()
plt.xticks(fontsize = 10,rotation = 90)

plt.show
```
