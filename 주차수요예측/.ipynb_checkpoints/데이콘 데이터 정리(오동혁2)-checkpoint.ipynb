<h4> 위 null값의 수를 확인하면 임대보증금, 임대료, 도보 10분 거리 내 지하철역 수(환승노선 수 반영),도보 10분거리 내 버스정류장 수 에 결측치가 존재하는 것을 확인 할 수 있다.</h4?

```python
data_train[['도보 10분거리 내 지하철역 수(환승노선 수 반영)', '도보 10분거리 내 버스정류장 수']] = data_train[['도보 10분거리 내 지하철역 수(환승노선 수 반영)', '도보 10분거리 내 버스정류장 수']].fillna(0)
#도보 10분거리 내 지하철역 수(환승노선 수 반영)-도보 10분거리 내 버스정류장 수 결측치 0으로 변화
# 그 이유는 광주지역은 지하철이 있어도 주변에 지하철이 있는 데이터가 없다 그래서 우리가 있고 없고를 정하기 힘들어 0으로 하는것이 좋을거라고 생각
```

```python
data_train[['임대보증금','임대료']] = data_train[['임대보증금','임대료']].replace('-', np.nan)
```

<h4>저희는 이 결측치를 값의 평균으로 채우려고 하는데 각 지역마다 임대보증금과 임대료가 다르고 그 지역안에서는 비슷할것이라고 생각했다. 그래서 각 지역의 임대보증금과 임대료의 평균을 구하여 그 지역의 평균으로 결측치를 채우고자 합니다. 그 전에 각 지역마다 임대 보증금과 임대료가 어떻게 분포되어있는지 비슷하게 분포되어 평균을 넣어도 괜찮을지 확인하기 위하여 각지역의 임대보증금, 임대료의 분포를 boxplot으로 확인 하여고 합니다.</h4> 

```python
tenancy_bond =  data_dynamic.pivot_table(data_dynamic, index = ['지역','임대보증금'])
local_tenancy_bond = tenancy_bond.reset_index('임대보증금').drop(['임대료','전용면적','전용면적별세대수'],axis=1)
```

```python
local_tenancy_bond
```

```python
sns.boxplot(x = local_tenancy_bond.index, y = local_tenancy_bond['임대보증금'])
plt.xticks(rotation = 45)
plt.show()
```

<h4>서울 데이터 빼고는 결측치가 있어도 대부분의 데이터가 그렇게 크게 분산 되어 있지 않아서 평균을 결측값으로 넣어도 괜찮다고 생각 합니다. 이그래프를 보고 역시 서울은 다르구나 라는 걸 느꼈습니다.</h4>

```python
data_dynamic_mean =  data_dynamic.groupby('지역')['임대보증금'].mean()
data_dynamic_mean
```

<h4>이제 각각의 지역마다 평균을 구해 각 지역의 nan값에 평균 값을 넣어서 얼마나 차이가 생기는 지 보겠습니다.</h4>

```python
local =list(data_dynamic_mean.index)
```

```python
dic = {}
for i in zip(local,data_dynamic_mean.values):
    dic[i[0]] = i[1]
```

```python
dic
```

```python
data_dynamic[data_dynamic['임대보증금'].isna()][['지역','임대보증금']]
```

<h4>각 지역의 평균 값을 계산하여 각 지역에 맞게 dictionary를 만든 다음에 for문으로 dic에 있는 값을 비교한 후 비어 있는 부분에 dic의 value값을 채워 넣으면 됩니다. </h4>

```python
for local in dic.keys():
    data_dynamic.loc[data_dynamic['지역'] == local, '임대보증금'] = data_dynamic.loc[data_dynamic['지역'] == local, '임대보증금'].fillna(dic[local])
```

```python
data_dynamic
```

<h4>이전에 보였던 data_dynamic에 있는 임대 보증금의 NaN 부분의 index에 있는 데이터를 다시 살펴보면 그 지역의 평균값으로 채워진것을 볼수 있습니다.</h4>
